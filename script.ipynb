```python
# import libraries
import pandas as pd
import numpy as np
from sklearn import ensemble, preprocessing, tree
```

```python
# load dataset
train = pd.read_csv('input/ford_train.csv')
test = pd.read_csv('input/ford_test.csv')
sample = pd.read_csv('input/example_submission.csv')
solution = pd.read_csv('input/solution.csv')
sample.Prediction.fillna
train.head()
```

```python
labels = train['IsAlert'].values
```

```python
train = train.drop('IsAlert', axis=1)
```

```python
# find irrelevant columns and drop them
print(train.P8.value_counts())
print(train.V7.value_counts())
train = train.drop('TrialID', axis=1)
train = train.drop('ObsNum', axis=1)
train = train.drop('P8', axis=1)
train = train.drop('V7', axis=1)
```

```python
train.head()
```

```python
# decision tree one (to find most important features)
my_tree_one = tree.DecisionTreeClassifier()
my_tree_one = my_tree_one.fit(train, labels)
```

```python
print(my_tree_one.feature_importances_)
```

```python
# Most important features: E9, V11, V1, P5, E7
train_features_five = train[['E9', 'V11', 'V1', 'P5', 'E7']].copy()
train_features_five.head()
```

```python
# decision tree classifier
my_tree_two = tree.DecisionTreeClassifier()
my_tree_two = my_tree_two.fit(train_features_five, labels)
```

```python
test.drop('IsAlert', axis=1)
test_features_five = test[['E9', 'V11', 'V1', 'E10', 'P7']].copy()
test_features_five.head()
```

```python
# create predictions and submission file
predictions = my_tree_two.predict_proba(test_features_five)[:, 1]
sample['Prediction'] = predictions
sample.to_csv('predictions.csv', index=False)
```

```python
# check with solution
solution_pred = solution.Prediction.values
predictions_int = predictions.astype(int)
```

```python
accuracy_decision_tree = (np.sum(predictions_int == solution_pred)) / (predictions_int.size)
```
